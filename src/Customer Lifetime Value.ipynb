```python
# Create an empty list to store data
outliers_info = []

for col in numeric_col.columns:
    #Calculated Q1, Q2, dan IQR
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    #Calculates upper and lower limits
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    #Identify outliers
    outliers_condition = (df[col] < lower_bound) | (df[col] > upper_bound)
    outliers = df[outliers_condition]

    #Added iteration results to outliers_info
    outliers_info.append([col, round(lower_bound, 2), round(upper_bound, 2), outliers_condition.sum()])

#Turning outliers_info into a dataframe
df_outliers = pd.DataFrame(outliers_info, columns=['Columns', 'Lower Bound', 'Upper Bound', 'Numbers of Outliers']).set_index('Columns').reset_index()
df_outliers
```

**Categorical Feature**

```python
# Menyimpan hasil ANOVA
categorical_anova_test = []

# Menghitung ANOVA untuk setiap variabel kategorikal
for var in categorical_col:
    # Kelompokkan Customer Lifetime Value berdasarkan kategori variabel
    groups = [df[df[var] == category]['Customer Lifetime Value'] for category in df[var].unique()]
    
    # Lakukan uji ANOVA
    anova_result = f_oneway(*groups)
    
    # Tentukan hipotesis
    p_value = anova_result.pvalue
    if p_value < 0.05:
        hypothesis = "There are Significant Difference"
    else:
        hypothesis = "There are No Significant Difference"
    
    # Simpan hasil
    categorical_anova_test.append({
        'Variable': var,
        'F-Statistics': anova_result.statistic,
        'p-value': p_value,
        'Hipotesis': hypothesis
    })

# Membuat DataFrame dari hasil ANOVA
df_categorical_anova_test = pd.DataFrame(categorical_anova_test)
df_categorical_anova_test
```

```python
# Create function metrics bussiness
def loss_cost(y_true, y_pred, cac=800, retention_cost=300, overprediction_weight=0.75, underprediction_weight=1.5):
    # Overspend on Acquisition: The difference between predicted CLV and actual CLV
    over = np.where(y_pred > y_true, (y_pred - y_true) * cac * overprediction_weight, 0)
    
    # Disadvantages of Underprediction: The difference between the actual CLV and the predicted CLV
    under = np.where(y_pred < y_true, (y_true - y_pred) * retention_cost * underprediction_weight, 0)

    # Sum over dan under
    return np.sum([over, under])

# Create metric loss
loss = make_scorer(score_func=loss_cost, greater_is_better=False)
```

The **loss_cost** function is designed to evaluate the financial impact of CLV prediction errors on an auto insurance company. It helps the company understand the potential losses resulting from inaccurate CLV predictions, enabling them to optimize their customer acquisition and retention strategies.

**Function Parameters:**
- y_true (array numpy):
    
    The actual value of the customer's CLV. This is the true CLV value obtained from historical data or direct observation.

- y_pred (array numpy):
    
    The predicted CLV value generated by a machine learning model or other predictive method.

- cac (biaya akuisisi pelanggan):
    
    The average cost of acquiring a new customer, set at $800 in this case. This includes marketing costs, sales costs, and other related expenses.

- retention_cost (biaya retensi per tahun):
    
    The average cost of retaining an existing customer for one year, set at $300 in this case.

- overprediction_weight:
    
    The weight for the loss caused by overprediction (overestimation of CLV). In this example, the weight is set to 0.75 times the acquisition cost ($600). This reflects the additional loss incurred by the company due to over-budgeting for acquisition.

- underprediction_weight:
    
    The weight for the loss caused by underprediction (underestimation of CLV). In this example, the weight is set to 1.5 times the retention cost ($450). This reflects the loss due to under-investment in customer retention that should have been higher.

```python
# Build class for data Feature Creation process
class FeatureCreation(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass

    def fit(self, X, y=None):
        return self
    
    def transform(self, X, y=None):
        # Feature creation
        X['Average Claim per Policy'] = X['Total Claim Amount'] / X['Number of Policies']
        X['Policy Interaction'] = X['Number of Policies'] * X['Renew Offer Type'].astype('category').cat.codes

        return X

# Build class for Detection Outliers
class OutlierDetection(BaseEstimator, TransformerMixin):
    def __init__(self, lower_percentile=10, upper_percentile=90):
        self.lower_percentile = lower_percentile
        self.upper_percentile = upper_percentile
        self.lower_bound = None
        self.upper_bound = None

    def fit(self, X, y=None):
        X = np.asarray(X)
        self.lower_bound = np.percentile(X, self.lower_percentile)
        self.upper_bound = np.percentile(X, self.upper_percentile)
        return self

    def transform(self, X):
        X = np.asarray(X)
        X = np.where(X < self.lower_bound, self.lower_bound, X)
        X = np.where(X > self.upper_bound, self.upper_bound, X)
        return X
    
# Build class for Selector Columns
class ColumnSelector(BaseEstimator, TransformerMixin):
    def __init__(self, columns=None):
        self.columns = columns
    
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        return X
    
# Ordinal mapping
ordinal_mapping = [
    {'col':'Coverage', 'mapping':{'Basic':1, 'Extended':2, 'Premium':3, None:0}},
    {'col':'Education', 'mapping':{'High School or Below':1, 'College':2, 'Bachelor':3, 'Master':4, 'Doctor':5, None:0}},
    {'col':'Renew Offer Type', 'mapping':{'Offer1':1, 'Offer2':2, 'Offer3':3, 'Offer4':4, None:0}}
]

# Pipeline Feature Engineering
robust_preprocessor_outlier = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('outlier', OutlierDetection()),
    ('scaler', RobustScaler())])

robust_preprocessor = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', RobustScaler())])

onehot_preprocessor = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('onehot', OneHotEncoder(handle_unknown='ignore', drop='first'))])

ordinal_preprocessor = Pipeline(steps=[
    ('ordinal', OrdinalEncoder(mapping=ordinal_mapping))])

feature_preprocessor = Pipeline(steps=[
    ('selectfeature', ColumnSelector())])


# Transformer from Pipeline Feature Engineering
# Preprocessor 1 for Set Feature 1
preprocessor1 = ColumnTransformer(
    transformers=[
        ("onehot", onehot_preprocessor, ['Vehicle Class', 'EmploymentStatus', 'Marital Status']),
        ("ordinal", ordinal_preprocessor, ['Coverage', 'Renew Offer Type', 'Education']),
        ("robust", robust_preprocessor, ['Income']),
        ("robust_outlier", robust_preprocessor_outlier, ['Monthly Premium Auto', 'Total Claim Amount']),
        ("colomselect",feature_preprocessor, ['Number of Policies'])
    ],
    remainder='drop', verbose_feature_names_out=False
)

# Preprocessor 2 for Set Feature 2
preprocessor2 = ColumnTransformer(
    transformers=[
        ("onehot", onehot_preprocessor, ['Vehicle Class', 'EmploymentStatus', 'Marital Status']),
        ("ordinal", ordinal_preprocessor, ['Coverage', 'Renew Offer Type', 'Education']),
        ("robust", robust_preprocessor, ['Income']),
        ("robust_outlier", robust_preprocessor_outlier, ['Monthly Premium Auto', 'Total Claim Amount', 'Average Claim per Policy']),
        ("colomselect",feature_preprocessor, ['Number of Policies', 'Policy Interaction'])
    ],
    remainder='drop', verbose_feature_names_out=False
)
```

<h2 style='text-align:left; color:darkblue'><b> Modeling Experiment </b></h1>

In this Experiment, the Metric Evaluations used are Mean Absolute Error (MAE), and Mean Absolute Percentage Error (MAPE).
- **Mean Absolute Error (MAE)**
    
    **Mean Absolute Error (MAE)** is an evaluation metric used to measure the accuracy of regression models in making predictions. MAE provides an average of the absolute errors between the predicted values by the model and the actual values from the data.
    
    Formula MAE:
    $$ 
    \text{MAE} = \frac{1}{n} \sum_{i=1}^{n} | y_i - \hat{y}_i | 
    $$

    where:
    - $y_i$ is Actual value.
    - $\hat{y}_i$ is Predicted value.
    - $n$ is Total number of data.

    **Explanation of the MAE Formula**

    1. **Absolute Error Calculation**: For each data point $i$, calculate the absolute error between the actual value $y_i$ and the predicted value $\hat{y}_i$. This is done using $| y_i - \hat{y}_i |$, which yields a positive or zero value.

    2. **Average Error**: Sum all the absolute errors and divide by the total number of data points $n$ to obtain the average error.

- **Mean Absolute Percentage Error (MAPE)**

    **Mean Absolute Percentage Error (MAPE)** is an evaluation metric that measures the magnitude of a model's error in terms of percentage of the actual value. This provides a more interpretable representation of relative error.

    Formula MAPE:
    $$
    \text{MAPE} = \frac{1}{n} \sum_{i=1}^{n} \left| \frac{y_i - \hat{y}_i}{y_i} \right| \times 100\%
    $$
    
    where:
    - $y_i$ is Actual value.
    - $\hat{y}_i$ is Predicted value.
    - $n$ is Total number of data.

    **Explanation of the MAPE Formula**

    1. **Relative Error Calculation**: For each data point $i$, calculate the relative error between the actual value $y_i$ and the predicted value $\hat{y}_i$. This is done using $\left| \frac{y_i - \hat{y}_i}{y_i} \right|$, which yields a positive value.

    2. **Average Relative Error**: Sum all the relative errors and divide by the total number of data points $n$ to obtain the average relative error.

    3. **Conversion to Percentage**: Multiply the result by 100% to express the value as a percentage.

**Create Function for Evaluation Model with Cross Validation**

```python
# Create Dict for Metric Evaluation
scoring = {
    'mae': 'neg_mean_absolute_error',
    'mape': 'neg_mean_absolute_percentage_error',
    'loss': loss
}

# Create function for Evaluation Model
def evaluate_models(models, pipeline, X_train, y_train, scoring):
    # Dict to save evaluation results
    results = {
        'Model': [],
        'mean_mae_train': [],
        'std_mae_train': [],
        'mean_mae_test': [],
        'std_mae_test': [],
        'mean_mape_train': [],
        'std_mape_train': [],
        'mean_mape_test': [],
        'std_mape_test': [],
        'mean_loss_train': [],
        'std_loss_train': [],
        'mean_loss_test': [],
        'std_loss_test': []
    }

    # Evaluate each model with the given pipeline
    for model, model_name in models.items():
        # Create a pipeline for the model
        pipeline = Pipeline(steps=[
            ('preprocessor', preprocessor1),
            ('model', model)])
        
        # Cross-validation to evaluate pipelines with different models
        cv_score = cross_validate(
            estimator=pipeline,
            X=X_train,
            y=y_train,
            scoring=scoring,
            cv=5,
            return_train_score=True
        )
        
        # Collect evaluation results for each metric
        results['Model'].append(model_name)
        
        # Mean and std for MAPE
        results['mean_mae_train'].append(-np.mean(cv_score['train_mae']))
        results['std_mae_train'].append(np.std(cv_score['train_mae']))
        results['mean_mae_test'].append(-np.mean(cv_score['test_mae']))
        results['std_mae_test'].append(np.std(cv_score['test_mae']))

        # Mean and std for MAPE
        results['mean_mape_train'].append(-np.mean(cv_score['train_mape']))
        results['std_mape_train'].append(np.std(cv_score['train_mape']))
        results['mean_mape_test'].append(-np.mean(cv_score['test_mape']))
        results['std_mape_test'].append(np.std(cv_score['test_mape']))

        # Mean and std for Loss
        results['mean_loss_train'].append(-np.mean(cv_score['train_loss']))
        results['std_loss_train'].append(np.std(cv_score['train_loss']))
        results['mean_loss_test'].append(-np.mean(cv_score['test_loss']))
        results['std_loss_test'].append(np.std(cv_score['test_loss']))

    # Create a DataFrame from the evaluation results
    results_df = pd.DataFrame(results)
    
    return results_df
```

Creating a function for Cross Validation to calculate training and validation scores. The scoring will utilize Mean Absolute Error (MAE) and Mean Absolute Percentage Error (MAPE), in addition to the custom **loss** function to evaluate model performance.

**Create Function to Randomized Search CV**

```python
def randomized_search(pipeline, parameters, scoring, X_train, y_train):
    # Menentukan RandomizedSearchCV dengan pipeline dan parameter distribusi
    randomized_search = RandomizedSearchCV(
        estimator=pipeline,
        param_distributions=parameters,
        n_iter=5,
        refit='mae',
        scoring=scoring,
        cv=5,
        random_state=42,
        return_train_score=True
    )
    
    # Melatih RandomizedSearchCV dengan data pelatihan
    randomized_search.fit(X_train, y_train)
    
    return randomized_search
```

Creating a function to perform Hyperparameter Tuning using RandomizedSearchCV. The scoring will utilize Mean Absolute Error (MAE) and Mean Absolute Percentage Error (MAPE), in addition to the custom **loss** function to evaluate model performance. The hyperparameter tuning process will be iterated 5 times, with 5-fold cross-validation performed at each iteration.

**Create Function to View Result of cv_result_**

```python
# Create list metric evaluation
metrics = ['mae', 'mape', 'loss']

# Create function
def cv_result(cv_results, metrics):
    # Konversi hasil cv_results ke DataFrame
    cv_results_df = pd.DataFrame(cv_results)
    
    results = {}
    
    for metric in metrics:
        train_mean_col = f'mean_train_{metric}'
        train_std_col = f'std_train_{metric}'
        test_mean_col = f'mean_test_{metric}'
        test_std_col = f'std_test_{metric}'
        
        # Mean and std for metrik train
        results[f'mean_{metric}_train'] = [-cv_results_df[train_mean_col].mean()] if metric in ['mae', 'mape', 'loss'] else [cv_results_df[train_mean_col].mean()]
        results[f'std_{metric}_train'] = [cv_results_df[train_std_col].mean()]
        
        # Mean and std for metrik test
        results[f'mean_{metric}_test'] = [-cv_results_df[test_mean_col].mean()] if metric in ['mae', 'mape', 'loss'] else [cv_results_df[test_mean_col].mean()]
        results[f'std_{metric}_test'] = [cv_results_df[test_std_col].mean()]

    return pd.DataFrame(results)
```

Creating a function to store the cv_results_ or results from **RandomizedSearchCV** into a DataFrame, filtered to only include metrics MAE, MAPE, and Loss. The function will then calculate the average of the 5 iterations performed by **RandomizedSearchCV**.

<h4 style='text-align:left; color:cyan'><b> How the Final Model Works </b></h4>

**The final model utilizes the Decision Tree algorithm with the following parameters:**
- *criterion='poisson'* : This criterion is used to select splits that minimize Mean Squared Error (MSE) with adjustments introduced by Jerome Friedman. It is designed to enhance the stability and accuracy of the Decision Tree model in regression contexts.

- *splitter='best'* : The model selects the best split at each node, optimizing the quality of splits at each step. The alternative option is 'random', which selects splits randomly, but 'best' ensures more optimal splits.

- *max_depth=12* : The maximum depth limit of the tree. This helps prevent the tree from growing too deep and overfitting.

- *min_samples_split=5* : Nodes will only be split if there are at least 5 samples in the node, preventing splits with too few samples.

- *min_samples_leaf=10* : Each leaf must have at least 10 samples. This prevents leaves with very little data and helps avoid overfitting.

- *random_state=42* : Sets a random seed to ensure consistent results each time the model is trained.


**The Decision Tree algorithm is defined as the variable *model_final_evl* which will be used within the TransformedTargetRegressor:**
